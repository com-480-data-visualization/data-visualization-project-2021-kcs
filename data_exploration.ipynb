```python
all_track_ids = [track['track']['id'] for track in playlist_1_details['tracks']['items']]
```

```python
len(all_track_ids)
```

```python
all_artists_ids = []
for track in playlist_1_details['tracks']['items']:
    artists = [artist['id'] for artist in track['track']['artists']]
    all_artists_ids.extend(artists)
```

```python
all_artists = []
artist_id_to_name = {}
id_set = set()
for track in playlist_1_details['tracks']['items']:
    artists = track['track']['artists']
    filtered_artists = []
    for a in artists:
        if not(a['id'] in id_set):
            id_set.add(a['id'])
            artist_id_to_name[a['id']] = a['name']
            filtered_artists.append(a)
        
    all_artists.extend(filtered_artists)
```

```python
related_artists = sp.artist_related_artists(all_artists_ids[0])
```

```python
from tqdm.auto import tqdm
```

```python
related_id_map = {}
for a_id in all_artists_ids:
    related_artists = sp.artist_related_artists(a_id)
    related_id_map[a_id] = [a['id'] for a in related_artists['artists']]
    
```

```python
graph_edges = []
```

```python
for a_id in related_id_map:
    for related_a_id in related_id_map[a_id]:
        if related_a_id in all_artists_ids:
            graph_edges.append((a_id, related_a_id))
```

```python
graph_edges
```

```python
artists_graph = nx.Graph()
```

```python
artists_graph.add_nodes_from(all_artists_ids)
```

```python
artists_graph.add_edges_from(graph_edges)
```

```python
import matplotlib.pyplot as plt
```

```python
artists_name_graph = nx.relabel_nodes(artists_graph, artist_id_to_name)
```

```python
# Draw graph or artists relationship
plt.figure(figsize=(20, 40))
nx.draw(artists_name_graph, nx.spring_layout(artists_name_graph, k=0.2, iterations=20), with_labels=True)
```

```python
# We are now going to link artist by shared genre. 
```

```python
artist_id_to_genres
```

```python
artists_genre_graph = nx.Graph()
artists_genre_graph.add_nodes_from(all_artists_ids)
```

```python
for a_id_1 in artist_id_to_genres:
    for a_id_2 in artist_id_to_genres:
        genres_1 = set(artist_id_to_genres[a_id_1])
        genres_2 = set(artist_id_to_genres[a_id_2])
        
        shared_genres = genres_1.intersection(genres_2)
        
        if len(shared_genres) > 2:
            artists_genre_graph.add_edge(a_id_1, a_id_2)
```

```python
artists_genre_graph = nx.relabel_nodes(artists_genre_graph, artist_id_to_name)
# Draw graph or artists relationship
plt.figure(figsize=(20, 40))
nx.draw(artists_genre_graph, nx.spring_layout(artists_genre_graph, k=0.2, iterations=20), with_labels=True)
```

```python
# We are going to draw another genre graph keeping only the pop genres:
pop_graphz = nx.Graph()
pop_graph.add_nodes_from(all_artists_ids)
```

```python
for a_id_1 in artist_id_to_genres:
    for a_id_2 in artist_id_to_genres:
        genres_1 = set(artist_id_to_genres[a_id_1])
        genres_2 = set(artist_id_to_genres[a_id_2])
        
        shared_genres = genres_1.intersection(genres_2)
        
        
        if True in [('pop' in g) for g in shared_genres] and len(shared_genres) > 2:
            pop_graph.add_edge(a_id_1, a_id_2, weight=len(shared_genres))
```

```python
pop_graph = nx.relabel_nodes(pop_graph, artist_id_to_name)
```
